# Sequence Classification 
**Note**: Since it was taking extremely long to train the model as per the specifications, I capped the LSTM model to 50 epochs and RNN model to 25 epochs. 
## Generate Data

```python
def gen_data(num_generate):
  fin_seqs = []
  target_seqs = []
  for i in range(num_generate):
    seq_len = random.randrange(100,110)
    # seq_len = 100
    seq_len = seq_len - 1
    curr_seq = []
    target_check = []
    curr_seq.append('E')
    t1 = random.randrange(10, 20)
    t2 = random.randrange(33, 43)
    t3 = random.randrange(66, 76) 
    for i in range(seq_len - 1):
      if i == t1:
        prob = random.random()
        if prob > 0.5:
          curr_seq.append('X')
          target_check.append(0)
        else:
          curr_seq.append('Y')
          target_check.append(1)
      elif i == t2:
        prob = random.random()
        if prob > 0.5:
          curr_seq.append('X')
          target_check.append(0)
        else:
          curr_seq.append('Y')
          target_check.append(1)
      elif i == t3:
          prob = random.random()
          if prob > 0.5:
            curr_seq.append('X')
            target_check.append(0)
          else:
            curr_seq.append('Y')
            target_check.append(1)
      else:
        prob = random.random()
        if prob <= 0.25:
          curr_seq.append('a')
        elif prob > 0.25 and prob <= 0.5:
          curr_seq.append('b')
        elif prob > 0.5 and prob <= 0.75:
          curr_seq.append('c')
        else:
          curr_seq.append('d')
    curr_seq.append('B')
    fin_seqs.append(curr_seq)
    if target_check[0] == 1 and target_check[1] == 1 and target_check[2] == 1:
      target_seqs.append('C')
    elif target_check[0] == 0 and target_check[1] == 1 and target_check[2] == 1:
      target_seqs.append('U')
    elif target_check[0] == 1 and target_check[1] == 0 and target_check[2] == 1:
      target_seqs.append('A')
    elif target_check[0] == 1 and target_check[1] == 1 and target_check[2] == 0:
      target_seqs.append('B')
    elif target_check[0] == 1 and target_check[1] == 0 and target_check[2] == 0:
      target_seqs.append('V')
    elif target_check[0] == 0 and target_check[1] == 1 and target_check[2] == 0:
      target_seqs.append('S')
    elif target_check[0] == 0 and target_check[1] == 0 and target_check[2] == 1:
      target_seqs.append('R')
    elif target_check[0] == 0 and target_check[1] == 0 and target_check[2] == 0:
      target_seqs.append('Q')
  return fin_seqs, target_seqs   
```

```python
fin, fin_target = gen_data(2560)
test_input, test_target = gen_data(300)
print(fin[1])
print(fin_target)
print(test_input)
print(test_target)
```

## One Hot Conversion 

```python
#one hot conversion functions 
def one_hot_util(char):
  if char == 'E':
    return 0 
  if char == 'a':
    return 1 
  if char == 'b':
    return 2
  if char == 'c':
    return 3
  if char == 'd':
    return 4
  if char == 'X':
    return 5 
  if char == 'Y':
    return 6 
  if char == 'B':
    return 7
def one_hot_target(char):
  if char == 'Q':
    return 0 
  if char == 'R':
    return 1 
  if char == 'S':
    return 2
  if char == 'U':
    return 3
  if char == 'V':
    return 4
  if char == 'A':
    return 5 
  if char == 'B':
    return 6 
  if char == 'C':
    return 7  
def one_hot(char, target = 0):
  if target == 0:
    return [1 if i==one_hot_util(char) else 0 for i in range(8)]
  else:
    return [1 if i==one_hot_target(char) else 0 for i in range(8)]
print(one_hot('b'))
print(one_hot('R', target = 1))
```

```python
# one hot conversion of the sequences 
fin_onehot = []
#training
for j in range(2560):
  one_hot_encoding = []
  for i in range(len(fin[j])):
    one_hot_e = one_hot(fin[j][i])
    one_hot_encoding.append(one_hot_e)
  fin_onehot.append(one_hot_encoding)
print(np.shape(fin_onehot[0]),np.shape(fin_onehot[1]))
#testing 
test_onehot = []
for j in range(300):
  one_hot_encoding = []
  for i in range(len(test_input[j])):
    one_hot_e = one_hot(test_input[j][i])
    one_hot_encoding.append(one_hot_e)
  test_onehot.append(one_hot_encoding)
print(np.shape(test_onehot[0]),np.shape(test_onehot[1]))
fin_onehot_target = []
one_hot_encoding = []
#train target 
for j in range(2560):
  one_hot_e = one_hot(fin_target[j], target=1)
  one_hot_encoding.append(one_hot_e)
  fin_onehot_target.append(np.array(one_hot_e))
#test target
test_onehot_target = []
one_hot_encoding = []
for j in range(300):
  one_hot_e = one_hot(test_target[j], target=1)
  one_hot_encoding.append(one_hot_e)
  test_onehot_target.append(np.array(one_hot_e))
print(fin_target[0], fin_target[1])
print(fin_onehot_target)
print(len(fin_onehot[0]), len(fin_onehot[1])) 
print(len(test_onehot[0]), len(test_onehot[1]))
print(test_onehot_target)
```

### Padding Sequences 
With this we will have same dimensions. Easier for us to input this to a model. 

```python
#padding sequences
tensor_input  = []
for i in range(2560):
    tensor_input.append(torch.tensor(fin_onehot[i]))
train_input = torch.nn.utils.rnn.pad_sequence(tensor_input, batch_first=True)
#train_target 
train_target = []
for i in range(2560):
    train_target.append(torch.tensor(fin_onehot_target[i]))
tensor_input = []
for i in range(300):
    tensor_input.append(torch.tensor(test_onehot[i]))
test_input = torch.nn.utils.rnn.pad_sequence(tensor_input, batch_first=True)
test_target = []
for i in range(300):
    test_target.append(torch.tensor(test_onehot_target[i]))
# test_target = torch.nn.utils.rnn.pad_sequence(tensor_input, batch_first=True)
```

## Building LSTM Architecture 

```python
class LSTM_Seq_Class(nn.Module):
    def __init__(self, input_size=8, output_size=8, num_layers=1):
        super().__init__()
        self.out_layer_size = output_size
        self.lstm1 = nn.LSTM(input_size, 2, num_layers)
        self.lstm2 = nn.LSTM(2, 4, num_layers)
        self.lstm3 = nn.LSTM(4, output_size, num_layers)

        # self.linear = nn.Linear(hidden_layer_size, output_size)

        self.hidden_cell1 = (torch.zeros(num_layers, 1, 2),
                             torch.zeros(num_layers, 1, 2))
        self.hidden_cell2 = (torch.zeros(num_layers, 1, 4),
                             torch.zeros(num_layers, 1, 4))
        self.hidden_cell3 = (torch.zeros(num_layers, 1, output_size),
                             torch.zeros(num_layers, 1, output_size))

    def forward(self, input_seq):
        lstm_out1, self.hidden_cell1 = self.lstm1(input_seq.view(len(input_seq), 1, -1), self.hidden_cell1)
        lstm_out2, self.hidden_cell2 = self.lstm2(lstm_out1, self.hidden_cell2)
        lstm_out3, self.hidden_cell3 = self.lstm3(lstm_out2, self.hidden_cell3)
        predictions = lstm_out3.view(len(input_seq), -1)
        return predictions[-1]
```

Here, we custom initialise weights and biases as specified in the question. We also choose the learning rate (0.1), loss function (Cross Entropy) and the optimiser (SGD) here. 

```python
model = LSTM_Seq_Class()
model.lstm1.weight_ih_l0 = torch.nn.Parameter(torch.FloatTensor(np.random.uniform(low=-0.1, high=0.1, size=(4*2,8))))
model.lstm2.weight_ih_l1 = torch.nn.Parameter(torch.FloatTensor(np.random.uniform(low=-0.1, high=0.1, size=(4*4,2))))
model.lstm3.weight_ih_l2 = torch.nn.Parameter(torch.FloatTensor(np.random.uniform(low=-0.1, high=0.1, size=(4*8,4))))
model.lstm1.weight_hh_l0 = torch.nn.Parameter(torch.FloatTensor(np.random.uniform(low=-0.1, high=0.1, size=(4*2,2))))
model.lstm2.weight_hh_l1 = torch.nn.Parameter(torch.FloatTensor(np.random.uniform(low=-0.1, high=0.1, size=(4*4,4))))
model.lstm3.weight_hh_l2 = torch.nn.Parameter(torch.FloatTensor(np.random.uniform(low=-0.1, high=0.1, size=(4*8,8))))
# print(torch.FloatTensor(np.full((4*2,),-2)))
model.lstm1.bias_ih_l0 = torch.nn.Parameter(torch.FloatTensor(np.full((4*2,),-2)))
# model.lstm1.bias_hh_l0 = torch.nn.Parameter(torch.FloatTensor(np.full((4*2,),-2)))
model.lstm2.bias_ih_l1 = torch.nn.Parameter(torch.FloatTensor(np.full((4*4,),-4)))
# model.lstm2.bias_hh_l1 = torch.nn.Parameter(torch.FloatTensor(np.full((4*4,),-4)))
model.lstm3.bias_ih_l2 = torch.nn.Parameter(torch.FloatTensor(np.full((4*8,),-6)))
# model.lstm3.bias_hh_l2 = torch.nn.Parameter(torch.FloatTensor(np.full((4*8,),-6)))

loss_function = nn.CrossEntropyLoss()
seq_classifier_loss = nn.L1Loss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)
print(model)
```

We capped the model at 50 epochs. But if you uncomment some parts of the code, you will be able to reach right training criteria. Due to lack of computational power I couldn't reach this point. We also plot the epoch vs loss curve of this model on training data. For the current number of epochs though, the loss was oscillating.  

```python
epochs = 50
seq_correct_count = 0
single_loss = loss_function(((torch.FloatTensor([0,0,0,0,1,0,0,0])).reshape(1,-1)),torch.LongTensor([2])) #just initialisation nothing more 
count = 0
flag = 0
i = 0
loss_epoch = []
index = []
# while single_loss.item() > 1 or seq_correct_count < 2 :
#     i += 1 
#     if flag == 1:
#       break 
for i in range(epochs):
    index.append(i)
    if flag == 1:
      break
    for seq, label in zip(train_input, train_target):
        optimizer.zero_grad()
        model.hidden_cell1 = (torch.zeros(1, 1, 2),
                              torch.zeros(1, 1, 2))
        model.hidden_cell2 = (torch.zeros(1, 1, 4),
                              torch.zeros(1, 1, 4))
        model.hidden_cell3 = (torch.zeros(1, 1, 8),
                              torch.zeros(1, 1, 8))
        seq = seq.float()
        label = label.float()
        seq, label = Variable(seq), Variable(label)
        _, target = label.max(dim=0) #changing from one hot target to class ID target. Cross entropy takes inputs like this. 
        y_pred = model(seq)
        threshold = seq_classifier_loss(y_pred, label)
        target = torch.LongTensor([target])
        single_loss = loss_function(y_pred.reshape(1,-1),target)
        if single_loss.item() <= 0.1 and seq_correct_count >= 2000:
          flag = 1 
          break 
        if threshold < 0.3:
          seq_correct_count +=1
        else:
          seq_correct_count = 0
        single_loss.backward()
        optimizer.step()
    loss_epoch.append(single_loss.item())
    # if i%2 == 1:
    print(f'epoch: {i:3} loss: {single_loss.item():10.8f}')
plt.plot(index, loss_epoch)
```

## Building RNN Architecture 

```python
#build architecture 
class RNN_Seq_Class(nn.Module):
    def __init__(self, input_size=8, output_size=8, num_layers=1):
        super().__init__()
        # self.hidden_layer_size = hidden_layer_size
        self.out_layer_size = output_size
        self.rnn1 = nn.RNN(input_size, 2, num_layers)
        # self.rnn1.weight_ih_l0 = torch.randn((2, input_size))
        # self.rnn1.weight_ih_l[0] = 2*torch.rand((2, 8)) - 1
        self.rnn2 = nn.RNN(2, 4, num_layers)
        self.rnn3 = nn.RNN(4, output_size, num_layers)

        # self.linear = nn.Linear(hidden_layer_size, output_size)

        self.hidden_cell1 = (torch.zeros(num_layers, 1, 2))
        self.hidden_cell2 = (torch.zeros(num_layers, 1, 4))
        self.hidden_cell3 = (torch.zeros(num_layers, 1, output_size))

    def forward(self, input_seq):
        rnn_out1, self.hidden_cell1 = self.rnn1(input_seq.view(len(input_seq), 1, -1), self.hidden_cell1)
        rnn_out2, self.hidden_cell2 = self.rnn2(rnn_out1, self.hidden_cell2)
        rnn_out3, self.hidden_cell3 = self.rnn3(rnn_out2, self.hidden_cell3)
        predictions = rnn_out3.view(len(input_seq), -1)
        return predictions[-1]
```

Here, we custom intialise the weights and the biases of the model. We also choose the loss function (CE), optimiser (SGD) and the learning rate (0.1)

```python
model = RNN_Seq_Class()
model.rnn1.weight_ih_l0 = torch.nn.Parameter(torch.FloatTensor(np.random.uniform(low=-0.1, high=0.1, size=(2,8))))
model.rnn2.weight_ih_l1 = torch.nn.Parameter(torch.FloatTensor(np.random.uniform(low=-0.1, high=0.1, size=(4,2))))
model.rnn3.weight_ih_l2 = torch.nn.Parameter(torch.FloatTensor(np.random.uniform(low=-0.1, high=0.1, size=(8,4))))
model.rnn1.weight_hh_l0 = torch.nn.Parameter(torch.FloatTensor(np.random.uniform(low=-0.1, high=0.1, size=(2,2))))
model.rnn2.weight_hh_l1 = torch.nn.Parameter(torch.FloatTensor(np.random.uniform(low=-0.1, high=0.1, size=(4,4))))
model.rnn3.weight_hh_l2 = torch.nn.Parameter(torch.FloatTensor(np.random.uniform(low=-0.1, high=0.1, size=(8,8))))
# print(torch.FloatTensor(np.full((4*2,),-2)))
model.rnn1.bias_ih_l0 = torch.nn.Parameter(torch.FloatTensor([-2,-2]))
# model.rnn1.bias_hh_l0 = torch.nn.Parameter(torch.FloatTensor([-2,-2]))
model.rnn2.bias_ih_l1 = torch.nn.Parameter(torch.FloatTensor([-4,-4,-4,-4]))
# model.rnn2.bias_hh_l1 = torch.nn.Parameter(torch.FloatTensor([-4,-4,-4,-4]))
model.rnn3.bias_ih_l2 = torch.nn.Parameter(torch.FloatTensor([-6,-6,-6,-6,-6,-6,-6,-6]))
# model.rnn3.bias_hh_l2 = torch.nn.Parameter(torch.FloatTensor([-6,-6,-6,-6,-6,-6,-6,-6]))
loss_function = nn.CrossEntropyLoss()
seq_classifier_loss = nn.L1Loss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)
print(model)
```
